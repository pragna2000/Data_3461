2. Loops

```python
def model_loop (x,w,b): 
    n= len(x)
    y = np.zeros(n)
    for i in range(n):
        y[i] = np.dot(w,x[i])+b
        
    return y


def model_np(x,w,b):
    return np.dot(x,w)+b


x = np.array([3, 6, 9, 12, 15])
w = 4
b = 21

model_loop(x, w, b)
model_np(x, w, b)
```

4. Create a function cost(x,y,w,b) that computes the mean square error between model prediction and data. 

```python
def cost(x, y, w, b):
    m=len(y)
    predictions = w * x + b
    squared_error = np.mean(predictions - y) ** 2
    return squared_error

#samplpe data
samples = 1000
x = np.linspace(-10, 20, samples)

sam_w = 4  
sam_b = 21 

noise = np.random.normal(0, 1, samples) 
y = sam_w * x + sam_b + noise

# y range
w_range = np.linspace(0, 4, 50)
b_range = np.linspace(-2, 2, 50)

cost_w = [cost(x,y,w,sam_b) for w in w_range]
cost_b = [cost(x,y,sam_w,b) for b in b_range]


```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(w_range, cost_w)
plt.title('Cost vs w (b held constant)')
plt.xlabel('w')
plt.ylabel('Cost (MSE)')
```

```python

plt.plot(b_range, cost_b)
plt.title('Cost vs b (w held constant)')
plt.xlabel('b')
plt.ylabel('Cost (MSE)')
```

5. Create a function dj_dw, dj_db = gradient(x, y, w, b)

```python
def gradient(x,y,w,b):
    
    y_pred = w * x + b
    data_w = -2 * np.mean((y - y_pred) * x)
    data_b = -2 * np.mean(y - y_pred)
    
    return data_w,data_b

x = np.array([2,4,7,10,13])
w_true = 2
b_true = 1
y_true = w_true*x+b_true


val_w = np.linspace(0,4,100)
val_b = np.linspace(-1,3,100)

data_w_val = [gradient(x,y_true,w,b_true) for w in val_w]
data_b_val = [gradient(x,y_true,w_true,b) for b in val_b]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(val_w, data_w_val, label='gradient vs. w', color='hotpink')
plt.axhline(y=0, color='purple', linestyle='-', linewidth=1) 
plt.xlabel('w')
plt.ylabel('gradient')
plt.title('gradient vs. w (holding b)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(val_b, data_b_val, label='gradient vs. w', color='purple')
plt.axhline(y=0, color='hotpink', linestyle='-', linewidth=1)
plt.xlabel('b')
plt.ylabel('gradient')
plt.title('gradient vs. b (holding w')
plt.legend()
plt.show()
```

6. Use the following gradient_descent function to "fit" a line to the data. Make sure you do not start at the optimal values from exercise 1. You may have to try different values of alpha.

```python
import copy
import math

def cost(x, y, w, b):
    y_pred = w * x + b
    mse = np.mean((y - y_pred) ** 2)
    return mse

def gradient(x, y, w, b):
    y_pred = w * x + b
    dx_dw = -2 * np.mean((y - y_pred) * x)
    dy_db = -2 * np.mean(y - y_pred)
    return dx_dw, dy_db

x = np.array([2, 4, 7, 10, 13])
w_true = 2
b_true = 1
y_true = w_true * x + b_true

def gradient_descent(x, y, w_in, b_in, alpha, iterations, cost, gradient): 
    w = copy.deepcopy(w_in) 
    
    x_history = []
    y_history = []

    b = b_in
    w = w_in
    
    for i in range(iterations):
        dx_dw, dy_db = gradient(x, y, w , b)     

        b = b - alpha * dy_db                            
        w = w - alpha * dx_dw                            

        if i < 100000:    
            J_history.append(cost(x, y, w, b))
            p_history.append([w, b])
            
        if i % math.ceil(iterations / 10) == 0:
            print(f"Iteration {i:4}: Cost {J_history[-1]:0.2e} ",
                  f"dx_dw: {dx_dw: 0.3e}, dy_db: {dy_db: 0.3e}  ",
                  f"w: {w: 0.3e}, b:{b: 0.5e}")
 
    return w, b, J_history, p_history 


w_initial = 0.8
b_initial = 0.4
alpha = 0.01  
num_iters = 100000 

w_optimal, b_optimal, x_history, y_history = gradient_descent(x, y_true, w_initial, b_initial, alpha, num_iters, cost, gradient)

print(f"Optimal w: {w_optimal}, Optimal b: {b_optimal}")
```
